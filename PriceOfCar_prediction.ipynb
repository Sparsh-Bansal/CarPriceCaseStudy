```python
def draw_bar_chart(col_name,fig_width,fig_height,xlabel='',ylabel='',title='',xLblRotation=0):
    df = master.groupby(master[col_name]).size().reset_index(name='Frequency')
    plt.figure(figsize=(fig_width,fig_height))
    sns.barplot(x=col_name,y='Frequency',data=df)
    plt.title(title,fontsize=50)
    plt.xticks(fontsize=30,rotation=xLblRotation)
    plt.yticks(fontsize=30)
    plt.xlabel(xlabel,fontsize=30)
    plt.ylabel(ylabel,fontsize=30)
    print(df)
    
#    plt.legend()
```

```python
def dummy_variable(data,variable,column_to_drop,Prefix=''):
    dummy=pd.get_dummies(data[variable],prefix=Prefix)
    data=pd.concat([data,dummy],axis=1)
    data=data.drop([variable,column_to_drop],axis=1)
    return data
```

```python
def draw_scatter_plot(data,X,Y,title):
    plt.figure(figsize=(20,10))
    plt.scatter(x=X,y=Y,s=np.pi*25,c=np.random.choice(['blue','green','red']),marker='o',alpha=0.75,data=data)
    plt.xlabel(X,fontsize=30)
    plt.ylabel('Price',fontsize=30)
    plt.title(title,fontsize=40)
    plt.xticks(fontsize=20)
    plt.yticks(fontsize=20)
    plt.legend(fontsize=20)
    plt.show()
    #plt.boxplot(x=data[X],y=data[Y],showfliers=True,fliersize=8,color='blue')
    
```

```python
def draw_boxplot(data,X):
    plt.figure(figsize=(15,5))
    sns.boxplot(x=data[X],color='blue',fliersize=10)
    plt.title('BoxPlot',fontsize=40)
    plt.xlabel(X,fontsize=30)
    plt.xticks(fontsize=20)
    plt.yticks(fontsize=20)
    plt.legend(fontsize=20)
    plt.show()
    print(master.shape)
```

```python
def remove_outliers(data,X):
    data= data[data[X] >= data[X].quantile(0.01)]
    data=data[data[X] <= data[X].quantile(0.99)]
    print(data.shape)
    return data
```

```python
from sklearn.model_selection import train_test_split as tt
X=master.drop(['price'],axis=1)
Y=master['price']
X_train,X_test,Y_train,Y_test= tt(X,Y,train_size=0.7,test_size=0.3,random_state=100)
```

```python
from sklearn.feature_selection import RFE
from sklearn.linear_model import LinearRegression
lm=LinearRegression()
lm.fit(X_train,Y_train)
rfe=RFE(lm,25)
rfe.fit(X_train,Y_train)
print(rfe.support_)
print(rfe.ranking_)
```

```python
col=X_train.columns[rfe.support_]
X_train_rfe=X_train[col]
X_test_rfe=X_test[col]
master_new=master[col]
master_new=pd.concat([master_new,master['price']],axis=1)
master_new.head()
```

```python
from statsmodels.stats.outliers_influence import variance_inflation_factor
import statsmodels.api as sm
X_train_rfe=sm.add_constant(X_train_rfe)
X_test_rfe = sm.add_constant(X_test_rfe)
# UDF for calculating vif value
def statsmodels_vif_cal(input_data, dependent_col):
    X = input_data.drop([dependent_col], axis=1)
    # For each X, calculate VIF and save in dataframe
    vif = pd.DataFrame(columns = ['Var', 'Vif'])
    vif["Var"] = X.columns
    vif["Vif"] = [variance_inflation_factor(X.values, i) for i in range(X.shape[1])]
    return vif.sort_values(by = 'Vif', axis=0, ascending=False, inplace=False).round(2)

statsmodels_vif_cal(master_new,'price')
# vif = pd.Series([variance_inflation_factor(X_train_rfe.values, i) for i in range(X_train_rfe.shape[1])])
# print(vif)
```
